```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
img = cv2.imread('perfectskeleton.tif',0)
plt.imshow(img, interpolation="sinc")
```

```python
kernelsize=np.ones((1,1))
skel=img.astype(float)
dilation=cv2.dilate(skel,kernelsize,iterations=1)
plt.imshow(dilation, cmap="Reds", interpolation='sinc')
cv2.imwrite("dilation1.tif", dilation)
```
