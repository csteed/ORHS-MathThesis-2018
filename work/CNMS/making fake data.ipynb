```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
img = cv2.imread('pathway.tif',0)
#plt.hist(img.ravel(),256,[0,256])
#plt.show()
(h,w)=np.shape(img)
centroid=np.zeros((h,w))
```

```python
blur = cv2.blur(img,(17, 17))
lc=blur*0.01
rangemax=np.amax(lc)
lc2=128-rangemax*0.5+lc
from scipy.stats import norm
noise = norm.rvs(centroid,12.75)
```

```python
ramp = np.multiply((np.ones((h,1))),(1+np.arange(w)))
ramp=(ramp/w)*128-64
#plt.imshow(ramp, cmap='gray')
lcn=lc2+noise+ramp
plt.imshow(lc2, cmap="gray")
```
